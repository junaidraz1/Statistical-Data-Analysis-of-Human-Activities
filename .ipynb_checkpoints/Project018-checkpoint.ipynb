```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import statistics as st
```

## Data preprocessing and cleaning

```python
# Load data and change the value to NaN when the value is not avalible.
columns = ['kohde', 'jasen', 'pvknro', 'sp', 'IKAL1', 'ASALUE', 'V1', 'V40', 'V7', 'V70', 'H1a_A', 'H1i_A']
df = pd.read_csv('habits.data', sep = ';', header = 0, usecols = columns, na_values = ['?'])
print(df, end = '\n\n')
```

#### Process missing value  
In order to do further analysis, we need to transform numeric columns to int64 type. So first we need to fill all NaN by 0.

```python
df.fillna(0, inplace=True)
for col in df:
    print(df[col].value_counts(), end = '\n\n')
```

#### Fix data inconsistency
From the data displayed above we can see some of these data is not useful.  
For example, some the value of v1 is standard time system, but anohter is not. So did we find from V7, V40, V70.  
So we need to transfrom the time to the same value.

```python
# Transform time data in the format of '00:00' to data in minutes
for col in ['V1', 'V7', 'V40', 'V70']:
    tmp = df[col].copy()
    for index in tmp.index:
        value = tmp.loc[index]
        if isinstance(value, str) and ':' in value:
            if len(value.split(':')) == 2:
                value += ':00'
            tmp.loc[index] = pd.to_timedelta(value).total_seconds() / 60
    print(tmp.value_counts(ascending=True), end = '\n\n')
    df[col] = tmp
```

For data in H1a_A and H1i_A, there are values other than 1 and 2.  
We assumed that it was caused by misunderstanding of the questionnaire by the respondent, who filled in time spent instead of yes or no.  
So we converted values more than 2 to 1, representing 'yes', and converted values of 0 to 2, representing 'no'.

```python
for col in ['H1a_A', 'H1i_A']:
    tmp = df[col].copy()
    tmp[tmp > 2] = 1
    tmp[tmp == 0] = 2
    print(tmp.value_counts(ascending=True), end = '\n\n')
    df[col] = tmp
```

```python
# Check datatype
print(df.dtypes, end = '\n\n')
```

#### Convert datatype
In order to conduct a more comprehensive analysis of the data, we converted the data type and filled the invalid values of the data.

```python
for col in ['V1', 'V7', 'V40', 'V70']:
    df[col] = df[col].apply(lambda x: int(float(x)) if isinstance(x, str) and '.' in x and x.replace('.', '', 1).isdigit() else x)
    df[col] = df[col].astype('int64')
df[['pvknro', 'sp', 'ASALUE', 'IKAL1', 'H1a_A', 'H1i_A']] = df[['pvknro', 'sp', 'ASALUE', 'IKAL1', 'H1a_A', 'H1i_A']].astype('category')

print(df.dtypes, end = '\n\n')
print(df, end = '\n\n')
```

```python
# Check the value category and confirm the value is useful.
print(pd.Categorical(df['H1a_A']), end = '\n\n')
print(pd.Categorical(df['H1i_A']), end = '\n\n')
```

#### Replace categorical column values
After that we replaced the values of categorical columns to understandable strings.

```python
categorical = {
    'pvknro': ('working_day', 'weekend'),
    'sp': ('male', 'female'),
    'ASALUE': ('city', 'municipality', 'rural_area'),
    'IKAL1': ('20-24', '25-34', '35-44', '45-54', 
              '55-64', '65-74', '75-'),
    'H1a_A': ('yes', 'no'),
    'H1i_A': ('yes', 'no')
}
for col in categorical:
    s = pd.Categorical(df[col]).rename_categories(categorical[col])
    df[col] = s
print(df, end = '\n\n')
```

#### Add new column for identifying respondents
We added a new column 'id' to identify each distinct respondent, which is a combination of 'kohde' and 'jasen'.

```python
df['id'] = df['kohde'].astype(str) + df['jasen'].astype(str)
print(df, end = '\n\n')
```

#### Make a new table for demographics
We also made a new table 'demo' for demographics, including 'id', 'sp', 'ASALUE', and 'IKAL1'.

```python
demo = df[['id', 'sp', 'ASALUE', 'IKAL1']].drop_duplicates().reset_index(drop = True)
print(demo, end = '\n\n')
```

#### Task 3: With respect to which activities do working days and weekends differ?
Before doing statistical analysis, we first examined the distribution of the activity duration data.

```python
# Made two datasets for working days and weekends
activity_working = activity[activity['pvknro'] == 'working_day']
activity_weekend = activity[activity['pvknro'] == 'weekend']

# Tested normality of V1, V40, V7, and V70 using histograms
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes[0, 0].hist(activity_working['V1'], bins = 50)
axes[1, 0].hist(activity_working['V40'], bins = 8)
axes[0, 1].hist(activity_working['V7'], bins = 8)
axes[1, 1].hist(activity_working['V70'], bins = 50)
axes[0, 0].set_title('V1')
axes[1, 0].set_title('V40')
axes[0, 1].set_title('V7')
axes[1, 1].set_title('V70')
fig.suptitle('Working Day')


fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes[0, 0].hist(activity_weekend['V1'], bins = 50)
axes[1, 0].hist(activity_weekend['V40'], bins = 8)
axes[0, 1].hist(activity_weekend['V7'], bins = 8)
axes[1, 1].hist(activity_weekend['V70'], bins = 50)
axes[0, 0].set_title('V1')
axes[1, 0].set_title('V40')
axes[0, 1].set_title('V7')
axes[1, 1].set_title('V70')
fig.suptitle('Weekend')

plt.show()
```

As shown in histograms above, we can find there are a large proportion of 0 in the dataset.  
It can be inferred that some respondents group do not do some specific activities at all.  
For example, some respondents do not work because they have retired or some other reasons.  
So we decided to omit these respondents when performing statistical tests.
#### Analysis Process
We analyzed each activity separately.  
We performed an inner join between activity_working and activity_weekend based on id, and then filtered out records with no activity time on both weekdays and weekends.  
For the filtered data, we conducted the Shapiro-Wilk test to check for normality.  
Based on the normality results, we selected appropriate statistical tests to compare the mean activity times between weekdays and weekends.

```python
# V1 Working
activity_merged_v1 = pd.merge(activity_working[['id', 'V1']], activity_weekend[['id', 'V1']], on = 'id', how = 'inner')
activity_merged_v1.columns = ['id', 'V1_working_day', 'V1_weekend']
activity_merged_v1 = activity_merged_v1[(activity_merged_v1['V1_working_day'] != 0) | (activity_merged_v1['V1_weekend'] != 0)].reset_index(drop = True)
print(activity_merged_v1, end = '\n\n')
print(stats.shapiro(activity_merged_v1.V1_working_day).pvalue, end = '\n\n') # shapiro test for v1_working_day
print(stats.shapiro(activity_merged_v1.V1_weekend).pvalue, end = '\n\n') # shapiro test for v1_weekend
print(stats.wilcoxon(activity_merged_v1.V1_working_day, activity_merged_v1.V1_weekend).pvalue, end = '\n\n') # wilcoxon signed-rank test
```

```python
# V7 Washing dishes
activity_merged_v7 = pd.merge(activity_working[['id', 'V7']], activity_weekend[['id', 'V7']], on = 'id', how = 'inner')
activity_merged_v7.columns = ['id', 'V7_working_day', 'V7_weekend']
activity_merged_v7 = activity_merged_v7[(activity_merged_v7['V7_working_day'] != 0) | (activity_merged_v7['V7_weekend'] != 0)].reset_index(drop = True)
print(activity_merged_v7, end = '\n\n')
print(stats.shapiro(activity_merged_v7.V7_working_day).pvalue, end = '\n\n') # shapiro test for v7_working_day
print(stats.shapiro(activity_merged_v7.V7_weekend).pvalue, end = '\n\n') # shapiro test for v7_weekend
print(stats.wilcoxon(activity_merged_v7.V7_working_day, activity_merged_v7.V7_weekend).pvalue, end = '\n\n') # wilcoxon signed-rank test
```

```python
# V40 Studing in school
activity_merged_v40 = pd.merge(activity_working[['id', 'V40']], activity_weekend[['id', 'V40']], on = 'id', how = 'inner')
activity_merged_v40.columns = ['id', 'V40_working_day', 'V40_weekend']
activity_merged_v40 = activity_merged_v40[(activity_merged_v40['V40_working_day'] != 0) | (activity_merged_v40['V40_weekend'] != 0)].reset_index(drop = True)
print(activity_merged_v40, end = '\n\n')
```

We found that there were only 5 records in the activity_merged_v40, which can not be performed any statistical analysis.  
So we can not get any significant results.

```python
# V70 Watching TV
activity_merged_v70 = pd.merge(activity_working[['id', 'V70']], activity_weekend[['id', 'V70']], on = 'id', how = 'inner')
activity_merged_v70.columns = ['id', 'V70_working_day', 'V70_weekend']
activity_merged_v70 = activity_merged_v70[(activity_merged_v70['V70_working_day'] != 0) | (activity_merged_v70['V70_weekend'] != 0)].reset_index(drop = True)
print(activity_merged_v70, end = '\n\n')
print(stats.shapiro(activity_merged_v70.V70_working_day).pvalue, end = '\n\n') # shapiro test for v70_working_day
print(stats.shapiro(activity_merged_v70.V70_weekend).pvalue, end = '\n\n') # shapiro test for v70_weekend
print(stats.wilcoxon(activity_merged_v70.V70_working_day, activity_merged_v70.V70_weekend).pvalue, end = '\n\n') # wilcoxon signed-rank test
```

#### Task 5: Which activities are associated with each other?

```python
activity_matrix = activity[['V1', 'V40', 'V7', 'V70']]

corr_matrix = pd.DataFrame(index = activity_matrix.columns, columns = activity_matrix.columns, dtype = float)
pval_matrix = pd.DataFrame(index = activity_matrix.columns, columns = activity_matrix.columns, dtype = float)
matrix = pd.DataFrame(index = activity_matrix.columns, columns = activity_matrix.columns)

for col1 in activity_matrix.columns:
    for col2 in activity_matrix.columns:
        if col1 == col2:
            corr_matrix.loc[col1, col2] = 1.0
            pval_matrix.loc[col1, col2] = 0.0
            matrix.loc[col1, col2] = '1.00 (p=0.000)'
        else:
            corr, pval = stats.spearmanr(activity_matrix[col1], activity_matrix[col2])
            corr_matrix.loc[col1, col2] = corr
            pval_matrix.loc[col1, col2] = pval
            matrix.loc[col1, col2] = f'{corr:.2f} (p={pval:.3f})'
print(matrix, end = '\n\n')

mask = pval_matrix > 0.05
corr_matrix_float = corr_matrix.astype(float)

plt.figure(figsize = (8, 6))
sns.heatmap(
    corr_matrix_float, 
    annot = True, 
    fmt = ".2f", 
    cmap = 'coolwarm', 
    mask = mask
)
plt.title('Correlation Matrix (p > 0.05 Masked)')
plt.show()
```

```python
# H1a_A and H1i_A
contingency_table = pd.crosstab(activity.H1a_A, activity.H1i_A)
print(f'H1a_A and H1i_A: {stats.chi2_contingency(contingency_table).pvalue}', end = '\n\n')

# H1a_A, H1i_A and V1, V7, V40, V70
columns = ['V1', 'V7', 'V40', 'V70']
for col in columns:
    activity_col  = activity[col]
    contingency_table = pd.crosstab(activity.H1a_A, activity_col)
    print(f'H1a_A and {col}: {stats.chi2_contingency(contingency_table).pvalue}', end = '\n\n')

for col in columns:
    activity_col  = activity[col]
    contingency_table = pd.crosstab(activity.H1i_A, activity_col)
    print(f'H1i_A and {col}: {stats.chi2_contingency(contingency_table).pvalue}', end = '\n\n')
```

#### Results for Task 5
As shown in the correlation matrix, heatmap, and p values above, we found significant negative correlations between V1 Working and V7 Washing dishes(***ρ = -0.17, p = 0.000***), as well as between V1 Working and V70 Watching TV(***ρ = -0.26, p = 0.000***). We also found H1a_A and H1i_A were associated with each other.
